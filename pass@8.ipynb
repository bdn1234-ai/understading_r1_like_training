```python
import pandas as pd
from math_verify import parse, verify

df = pd.read_csv("analysis_base_model/exploration_ability/Qwen2.5-Math-1.5B/Qwen2.5-Math-1.5B_1_pass8.csv")

def check_correctness(answer, ground_truth):
    try:
        parsed_answer = parse(str(answer))
        parsed_gt = parse(str(ground_truth))
        
        return verify(parsed_answer, parsed_gt)
    except Exception:
        return False

answer_cols = [f'Answer_{i}' for i in range(1, 9)]

correct_cols = []
for col in answer_cols:
    col_name = f'{col}_is_correct'
    df[col_name] = df.apply(lambda row: check_correctness(row[col], row['Ground_Truth']), axis=1)
    correct_cols.append(col_name)

df['is_pass_8'] = df[correct_cols].any(axis=1)

pass8_count = df['is_pass_8'].sum()
total_questions = len(df)

print(f"Pass@8: {(pass8_count / total_questions) * 100:.2f}%")
```
