```python
import pandas as pd

df = pd.read_csv("analysis_base_model/exploration_ability/Qwen2.5-Math-1.5B/Qwen2.5-Math-1.5B_1_pass8.csv")

# Giả sử bạn có 1 cột 'Ground_Truth' chứa đáp án đúng
# và bạn có 1 hàm đánh giá xem câu trả lời có đúng hay không
def check_correctness(answer, ground_truth):
    # Bạn sẽ cần tự định nghĩa logic chấm điểm ở đây.
    # Ví dụ đơn giản nhất là kiểm tra xem đáp án đúng có nằm trong câu trả lời không:
    return str(ground_truth).strip() in str(answer)

# 1. Lấy tên 8 cột
answer_cols = [f'Answer_{i}' for i in range(1, 9)]

# 2. Kiểm tra tính đúng đắn (Correctness) của từng câu trả lời
correct_cols = []
for col in answer_cols:
    col_name = f'{col}_is_correct'
    # Áp dụng hàm chấm điểm cho từng hàng
    df[col_name] = df.apply(lambda row: check_correctness(row[col], row['Ground_Truth']), axis=1)
    correct_cols.append(col_name)

# 3. Tính Pass@8: Có ÍT NHẤT 1 câu đúng trong 8 câu
df['is_pass_8'] = df[correct_cols].any(axis=1)

# 4. In thống kê
pass8_count = df['is_pass_8'].sum()
total_questions = len(df)

print(f"Số lượng câu hỏi Pass@8: {pass8_count} / {total_questions}")
print(f"Tỉ lệ Pass@8: {(pass8_count / total_questions) * 100:.2f}%")
```
