```python
import pandas as pd

def count_sr_keywords(file_name: str):
    try:
        df = pd.read_csv(file_name)
        output = df.to_dict('records')
    except Exception as e:
        print(f"Read file error: {e}")
        return


    keyword_pool = [
        "recheck", "rethink", "reassess", "reevaluate", "re-evaluate", 
        "reevaluation", "re-examine", "reexamine", "reconsider", "reanalyze",
        "double-check", "check again", "think again", "verify again", "go over the steps"
    ]

    sr_per_question_list = []
    sr_per_response_list = []

    sr_per_question = 0
    sr_per_response = 0

    for o in output:
        question_has_sr = False

        for i in range(1, 9):
            col_name = f"Answer_{i}"
            
            if col_name not in o or pd.isna(o[col_name]):
                continue

            response = str(o[col_name]).lower()

            keyword_detected = any(keyword in response for keyword in keyword_pool)

            if keyword_detected:
                sr_per_response += 1          
                question_has_sr = True        

        if question_has_sr:
            sr_per_question += 1              

        sr_per_response_list.append(sr_per_response)
        sr_per_question_list.append(sr_per_question)

    return {
        "response_level": sr_per_response_list,
        "question_level": sr_per_question_list
    }
```

```python
# Gọi hàm xử lý
FILE_NAME = "/Users/admin/understading_r1_like_training/res_analysis/exploration_ability/Qwen2.5-Math-1.5B/Qwen2.5-Math-1.5B_0.7_pass8.csv"
results = count_sr_keywords(FILE_NAME)
print(f"Total number of keywords appearing (Response-level): {results['response_level'][-1]}")
print(f"Number of questions containing keywords (Question-level): {results['question_level'][-1]}")
```
